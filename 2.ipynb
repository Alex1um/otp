```scilab
// 3.1
// [u, v] = meshgrid(linspace(0, 8*%pi, 80), linspace(0, 8*%pi, 80));
u = linspace(0, 8*%pi, 100);
v = linspace(0, 8*%pi, 100);
x = (cos(u).*u)'*(1 + cos(v / 2));
y = (u / 2)'*sin(v);
z = (sin(u).*u)'*(1 + cos(v / 2));
plot3d2(x, y, z)
```

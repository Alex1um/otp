```scilab
s = poly(0, "S")
h = 0.01
```

```scilab
Tkr = 12.8
Kkr = 1.18
Ti = Tkr / 2
Td = Ti / 4
Tc = Td / 8
T0 = 1.17
n = 6
w1 = (1 + 1 / ((Tkr / 1.2 - 6.5)*s)) * (Kkr * 0.45 + 0.068) * (2*exp(1.2)/(1 + T0 * s)^n)
w = w1 / (1 + w1)
```

```scilab
sys = syslin('c', w);
sysd = dscr(sys, h);
Ad = sysd.A
```

```scilab
I = eye(Ad);
H = lyap(Ad, -I, 'd')
s = spec(H)
```

```scilab
norm(s, 2)
```

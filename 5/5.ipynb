```scilab
s = poly(0, "s")
```

```scilab
Tkr = 12.8
Kkr = 1.18
Ti = Tkr / 2
Td = Ti / 4
Tc = Td / 8
T0 = 1.17
n = 6
w1 = (1 + 1 / ((Tkr / 1.2 - 6.5)*s)) * (Kkr * 0.45 + 0.068) * (2/(1 + T0 * s)^n)
w = w1 / (1 + w1)
```

```scilab
a = w.den
b = w.num
Sl = syslin('c', b, a)
```

```scilab
nyquist(Sl)
```

```scilab
bode(Sl)
```

```scilab
numer = coeff(w.num)
denom = coeff(w.den)
denom = denom(1:length(denom) - 1);
```

```scilab
A = [zeros(1, length(denom) - 1); eye(length(denom) - 1, length(denom) - 1)];
A = [A -denom']
```

```scilab
I = eye(A);
H = lyap(A, -I, 'c')
```

```scilab
spec(H)
spec(A)
```

```scilab
min(spec(H))
max(spec(H))
```
